```python
# coding:utf-8

import pandas as pd
import matplotlib.pyplot as plt
from qdata.baidu_index import get_search_index
# from qdata.baidu_index.common import check_keywords_exists
# from qdata.baidu_login import get_cookie_by_qr_login
from Util import read_json

cookie = read_json('config.json')['baidu_cookie']
keywords_list = [['发烧'],['旅游']]
```

```python
dt = get_search_index(keywords_list=keywords_list, start_date='2023-01-01', end_date='2024-01-10', cookies=cookie)
```

```python
df = pd.DataFrame(list(dt))
```

```python
df = df[df['type']=='all']
```

```python
df.head()
```

```python
df['num']=pd.to_numeric(df['index'])
df.reset_index(inplace=True)
```

```python
df.head()
```

```python
plt.figure(figsize=(9, 5))
plt.plot(df['date'], df['num'])
plt.xlabel('date')
plt.ylabel('indexes')
plt.savefig('baidu_index2.png', facecolor='w')
plt.close()
```

```python
df['dtm'] = pd.to_datetime(df['date'])
```

```python
plt.figure(figsize=(9, 9))
plt.plot(df['dtm'], df['num'])
plt.xlabel('date')
plt.ylabel('indexes')
plt.savefig('baidu_index1.png')
plt.close()
```
