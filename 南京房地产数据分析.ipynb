```python
import pandas as pd
import numpy as np

file1=r'D:\Programing\python_projects\Self-study\data_set\Xiaoqu_NJ.csv'
xiaoqu_raw=pd.read_csv(file1,header=None, names=["数据采集日期", "来源", "城市定位", "区县定位", "街道定位", "小区定位",
                                                 "地址", "标题", "热度", "参考均价", "建筑年代", "建筑类型", "物业费", 
                                                 "物业", "开发商", "楼栋总数", "房屋总数", "附近房产中介", "经度", "纬度"])
xiaoqu_raw
```

```python
# step 1: 删除关键信息为空的行
xiaoqu_raw=xiaoqu_raw[(xiaoqu_raw['小区定位'].notnull()) | (xiaoqu_raw['城市定位'].notnull())]
# step 2: 丢弃无关信息列
del xiaoqu_raw['数据采集日期']
del xiaoqu_raw['附近房产中介']
del xiaoqu_raw['来源']
del xiaoqu_raw['地址']
del xiaoqu_raw['标题']
xiaoqu_raw
```

```python
# step 3: 提取关键标签
xiaoqu_raw['城市定位']=xiaoqu_raw['城市定位'].apply(lambda x : str(x).replace('小区', ''))
xiaoqu_raw['区县定位']=xiaoqu_raw['区县定位'].apply(lambda x : str(x).replace('小区', ''))
xiaoqu_raw['街道定位']=xiaoqu_raw['街道定位'].apply(lambda x : str(x).replace('小区', ''))
xiaoqu_raw
```

```python
# step 4: 提取信息中的关键数字
import re
# 提取整数型
def parse_integer(x):
    # 防止异常值扰乱正常判断
    if isinstance(x, float) or isinstance(x, int):
        return np.nan
    pattern=re.compile(r'\d+')
    res=re.findall(pattern, x)
    # 返回值为包含匹配结果的列表，如果列表为空，则未匹配上；
    if len(res)==1:
        return res[0]
    else:
        return np.nan

# 提取参考均价中的单价数据
def parse_unitprice(x):
    if isinstance(x, float) or isinstance(x, int):
        return np.nan
    pattern=re.compile(r'(\d+\.{0,1}\d*)元/㎡')
    res=re.findall(pattern, x)
    if len(res)==1:
        return res[0]
    else:
        return np.nan
    
# 提取物业费中的价格信息
def parse_price(x):
    if isinstance(x, float) or isinstance(x, int):
        return np.nan
    pattern=re.compile(r'\d+\.{0,1}\d*')
    res=re.findall(pattern, x)
    if len(res)==1:
        return res[0]
    else:
        return np.nan
    
# 提取"热度"中关注的用户数，值越大，表示小数热度越高
xiaoqu_raw['热度']=xiaoqu_raw['热度'].apply(parse_integer)
# 提取建筑年份信息
xiaoqu_raw['建筑年代']=xiaoqu_raw['建筑年代'].apply(parse_integer)
# 提取房屋栋数
xiaoqu_raw['楼栋总数']=xiaoqu_raw['楼栋总数'].apply(parse_integer)
# 提取房屋总数
xiaoqu_raw['房屋总数']=xiaoqu_raw['房屋总数'].apply(parse_integer)
#提取小区均价，单位元/㎡
xiaoqu_raw['参考均价']=xiaoqu_raw['参考均价'].apply(parse_unitprice)
# 提取小区物业费，单位元/平米/月
xiaoqu_raw["物业费"] = xiaoqu_raw["物业费"].apply(parse_price)
# 计算房龄
xiaoqu_raw['房龄']=xiaoqu_raw['建筑年代'].apply(lambda x : np.nan if pd.isnull(x) else 2019-int(x))
```

```python
# 案例二：南京历年小区建成情况-折线图
# 计算南京历年的小区建成数量
jianzhu_tmp=xiaoqu['建筑年代'].value_counts()
jianzhu_year=jianzhu_tmp.sort_index().reset_index()
jianzhu_year
```

```python
# 时间跨度太大，因此选择80年以后、18年以前的数据
jianzhu_year.columns=['year', 'cnt']
jianzhu_year=jianzhu_year[(jianzhu_year['year']>='1980') & (jianzhu_year['year']<='2019')]

plt.figure(figsize=(16,6))
# 基于plot()语句（折线）
plt.plot(jianzhu_year['year'], jianzhu_year['cnt'], color='#A0522D', alpha=0.9, marker='o',
         label='小区数量', linewidth=2, linestyle='--')
plt.title('南京小区建成数量与年份关系', fontsize=15)
plt.legend()
```

```python
plt.figure(figsize=(16,6))
plt.plot(jianzhu_year['year'], jianzhu_year['cnt'], linestyle='-', alpha=0.9,
         linewidth=2, color='#A0522D', label='小区数量')
# 绘制数据的均线，并进行颜色填充，增加图形的可读性
plt.fill_between(jianzhu_year['year'], jianzhu_year['cnt'].mean(), jianzhu_year['cnt'],
                 color='#CD853F')
plt.title('南京小区建成数量与年份的关系', fontsize=15)
plt.legend()
```
